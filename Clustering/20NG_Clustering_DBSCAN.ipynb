```python
def create_neighbors(eps):
    neighbors_dict = {}
    n_count = {}
    for i in range(len(vectors)):
        neighbors_dict[i] = []
        n_count[i] = 0

    for i in range(len(vectors)):
        print(i)
        for j in range(len(vectors)):
            if(dist[i][j] <= eps):
                neighbors_dict[i] += list([j])
                n_count[i] += 1
#                 neighbors_dict[j] += list([i])        
                
                
    for i in neighbors_dict:
        neighbors_dict[i] = list(set(neighbors_dict[i])) 
    return neighbors_dict, n_count
```

```python
import itertools
import operator

def most_common(L):
    SL = sorted((x, i) for i, x in enumerate(L))
    groups = itertools.groupby(SL, key=operator.itemgetter(0))
    def _auxfun(g):
        item, iterable = g
        count = 0
        min_index = len(L)
        for _, where in iterable:
            count += 1
            min_index = min(min_index, where)
        return count, -min_index
    return max(groups, key=_auxfun)[0]
```

```python
def get_dict(cluster):
    d = dict()
    for j in cluster:
        label = test_Y[j]
        if label in d:
            d[label] += 1
        else:
            d[label] = 1
    return d
```

```python
def get_purity(clusters):
    size = []
    purity = []
    for cluster in clusters:
        print(type(clusters[cluster]))
        d = get_dict(clusters[cluster])
        l = len(clusters[cluster])
        size.append(l)
        purity.append(max(list(d.values())))

    return sum(purity)/2000
```
